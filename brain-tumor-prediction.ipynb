**Brain Tumor Prediction using YOLOv8 model**

**Importing Libraries**

```python
import ultralytics
ultralytics.checks()
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')              ##use 'yolov8' nano model

##model training
model.train(
    data = '/kaggle/input/brain-tumor-zip/data.yaml',
    epochs = 3,
    imgsz = 640,
    batch = 32,
    workers = 2 )

```

**Evaluate the model**

```python
metrics = model.val(save = True)
```

So our results are saved in 'run/detect/train32'


**Function to display images**

```python
import os
from IPython.display import Image, display

result_dir = '/kaggle/working/runs/detect/train32/'  # or 'train2' if you want training results
image_files = [f for f in os.listdir(result_dir) if f.endswith('.jpg')]

print("Available result images:", image_files)
```

**To disply the first predicted image**

```python
##TO display the first image
img_path = os.path.join(result_dir, image_files[0])  # pick the first one
display(Image(filename=img_path))
```
